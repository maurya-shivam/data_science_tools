```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('classic')
```

```python
def test_my_first_plot():
    fig = plt.figure(figsize = (8, 6))
    ax = fig.add_subplot(111)
    t = [x for x in range(5, 26, 5)]
    d = [x for x in range(25, 126, 25)]
    ax.plot(t, d, label = 'd = 5t')
    ax.set(xlabel = 'time (seconds)', ylabel = 'distance (meters)', title = 'Time vs Distance Covered', xlim = (0, 30), ylim = (0, 130))
    #ax.set_xlabel('time (seconds)'); ax.set_ylabel('distance (meters)'); ax.set_title('Time vs Distance Covered'); ax.set_xlim(0, 30); ax.set_ylim(0, 130)
    ax.legend(loc = 'upper left', fontsize = '12')
    #plt.savefig('scatter.png')
    
test_my_first_plot()
```

```python
def test_sine_wave_plot():
    fig = plt.figure(figsize = (12, 3))
    ax = fig.add_subplot()
    t = np.linspace(0.0, 2.0, 200); v = np.sin(2.5*np.pi*t)
    plt.plot(t, v, label = 'sin(t)', c = 'r')
    #ax.set(xlabel = 'Time (seconds)', ylabel = 'Voltage (mV)', title = 'Sine Wave', xlim = (0, 2), ylim = (-1, 1))
    plt.title('Sine Wave'); plt.xlabel('Time (seconds)'); plt.ylabel('Voltage (mV)'); plt.xlim(0, 2); plt.ylim(-1, 1);
    ax.xaxis.set_major_locator(plt.MultipleLocator(0.2))
    #ax.yaxis.set_major_locator(plt.MultipleLocator(1))
    ax.set_yticks([1, 0, -1])
    plt.grid(linestyle = '--')
    plt.legend()
    #plt.savefig('sinewave.png')
    
test_sine_wave_plot()
```

```python
def test_multi_curve_plot():
    fig = plt.figure(figsize = (12, 3))
    ax = fig.add_subplot()
    x = np.linspace(0, 5.0, 20)
    y1 = x; y2 = x**2; y3 = x**3
    plt.plot(x, y1, 'red', label = 'y = x', marker = 'o')
    plt.plot(x, y2, 'green', label = 'y = x**2', marker = 's')
    plt.plot(x, y3, 'blue', label = 'y = x**3', marker = '^')
    plt.title('Linear, Quadratic, & Cubic Equations'); plt.xlabel('x'); plt.ylabel('f(x)')
    plt.legend(loc = 'upper left', fontsize = 10)
    #plt.savefig('multicurve.png')
    
test_multi_curve_plot()
```

```python
def test_scatter_plot():
    fig = plt.figure(figsize = (12, 3))
    ax = fig.add_subplot()
    s = [50, 60, 55, 50, 70, 65, 75, 65, 80, 90, 93, 95]
    months = [x for x in range(1, 13)]
    ax.set(title = "Cars Sold by Company 'X' in 2017", xlabel = 'Months', ylabel = 'No. of Cars Sold', xlim = (0, 13), ylim = (20, 100))
    ax.scatter(months, s, c = 'red', marker = 'o')
    xticksvalues = [1,3,5,7,9,11]
    xticklabels = ['Jan', 'Mar', 'May', 'Jul', 'Sep', 'Nov']
    #ax.set_xticks([1,3,5,7,9,11])   #  plt.xticks(data)
    #ax.set_xticklabels(['Jan', 'Mar', 'May', 'Jul', 'Sep', 'Nov'])
    plt.xticks(xticksvalues, xticklabels)
    #plt.savefig('scatter.png')
    
test_scatter_plot()
```

```python
def test_barplot_of_iris_sepal_length():
    #ax.set(xlabel = 'Species', ylabel = 'Sepal Length (cm)', title = 'Mean Sepal Length of Iris Species', xlim = (0, 3), ylim = (0, 7))
    plt.title('Mean Sepal Length of Iris Species'); plt.xlabel('Species'); plt.ylabel('Sepal Length (cm)'); plt.xlim(0, 3); plt.ylim(0, 7);
    species = ['setosa', 'versicolor', 'viriginica']
    index = [0.2, 1.2, 2.2]
    sepal_length = [5.01, 5.94, 6.59]
    xticks = [0.45, 1.45, 2.45]
    xticklabels = species
    plt.xticks(xticks, xticklabels)
    plt.bar(index, sepal_length, width = 0.5, color = 'r', edgecolor = 'k' )

def test_hbarplot_of_iris_petal_length():
    #ax.set(ylabel = 'Species', xlabel = 'Petal Length (cm)', title = 'Mean Petal Length of Iris Species')
    plt.title('Mean Petal Length of Iris Species'); plt.ylabel('Species'); plt.xlabel('Petal Length (cm)');
    species = ['setosa', 'versicolor', 'viriginica']
    index = [0.2, 1.2, 2.2]
    petal_length = [1.46, 4.26, 5.55]
    yticks = [0.45, 1.45, 2.45]
    yticklabels = species
    plt.yticks(yticks, yticklabels)
    plt.barh(index, petal_length, color = 'c', edgecolor = 'k', height = 0.5 )

fig = plt.figure(figsize = (16,8))
ax = fig.add_subplot(221)
test_barplot_of_iris_sepal_length()
ax = fig.add_subplot(222)
test_hbarplot_of_iris_petal_length()
```

```python
def test_barplot_of_iris_measurements():
    fig = plt.figure(figsize = (7, 4))
    ax = fig.add_subplot(111)
    #ax.set(xlabel = 'Species', ylabel = 'Iris Measurements (cm)', title = 'Mean Measurements of Iris Species', xlim = (0.5, 3.7), ylim = (0, 10))
    plt.title('Mean Measurements of Iris Species'); plt.xlabel('Species'); plt.ylabel('Iris Measurements (cm)'); plt.xlim(0.5, 3.7); plt.ylim(0, 10);
    sepal_len = [5.01, 5.94, 6.59]; sepal_wd = [3.42, 2.77, 2.97]; petal_len = [1.46, 4.26, 5.55]; petal_wd = [0.24, 1.33, 2.03]
    species_index1 = [0.7, 1.7, 2.7]; species_index2 = [0.9, 1.9, 2.9]; species_index3 = [1.1, 2.1, 3.1]; species_index4 = [1.3, 2.3, 3.3]
    species = ['setosa', 'versicolor', 'viriginica']
    xticks = [1.1, 2.1, 3.1]
    xticklabels = species
    plt.xticks(xticks, xticklabels)
    plt.bar(species_index1, sepal_len, width = 0.2, color = 'c', edgecolor = 'k', label = 'Sepal Length')
    plt.bar(species_index2, sepal_wd, width = 0.2, color = 'm', edgecolor = 'k', label = 'Sepal Width')
    plt.bar(species_index3, petal_len, width = 0.2, color = 'y', edgecolor = 'k', label = 'Petal Length')
    plt.bar(species_index4, petal_wd, width = 0.2, color = 'orange', edgecolor = 'k', label = 'Petal Width')
    plt.legend(loc = 2, fontsize = 11)

test_barplot_of_iris_measurements()
```

```python
def test_boxplot_of_four_normal_distribution():
    fig = plt.figure(figsize = (8, 6))
    ax = fig.add_subplot(111)
    np.random.seed(100)
    x1 = np.random.normal(loc = 25, scale = 3.0, size = 1000); x2 = np.random.normal(loc = 35, scale = 5.0, size = 1000); x3 = np.random.normal(loc = 55, scale = 10.0, size = 1000); x4 = np.random.normal(loc = 45, scale = 3.0, size = 1000)
    labels = ['X1', 'X2', 'X3', 'X4']
    ax.set(title = 'Box plot of Multiple Datasets', xlabel = 'Dataset', ylabel = 'Value')
    plt.boxplot([x1, x2, x3, x4], labels = labels, notch = True, sym = '+', patch_artist = True)
    plt.show()
    
test_boxplot_of_four_normal_distribution()
```

```python
def test_generate_plot_with_style(style, count):
    with plt.style.context(style):
        fig = plt.figure(figsize = (8, 6))
        ax = fig.add_subplot(111)
        sepal_len = [5.01, 5.94, 6.59]; sepal_wd = [3.42, 2.77, 2.97]; petal_len = [1.46, 4.26, 5.55]; petal_wd = [0.24, 1.33, 2.03]
        species = ['setosa', 'versicolor', 'viriginica']
        species_index1 = [0.7, 1.7, 2.7]; species_index2 = [0.9, 1.9, 2.9]; species_index3 = [1.1, 2.1, 3.1]; species_index4 = [1.3, 2.3, 3.3]
        ax.set(title = 'Mean Measurements of Iris Species', xlabel = 'Species', ylabel = 'Iris Measurements (cm)', xlim = (0.5, 3.7), ylim = (0, 10))
        plt.xticks([1.1, 2.1, 3.1], species)
        plt.bar(species_index1, sepal_len, label = 'Sepal Length', width = 0.2)
        plt.bar(species_index2, sepal_wd, label = 'Sepal Width', width = 0.2)
        plt.bar(species_index3, petal_len, label = 'Petal Length', width = 0.2)
        plt.bar(species_index4, petal_wd, label = 'Petal Width', width = 0.2)
        plt.legend()
        plt.savefig('plotstyle{}.png'.format(count))
        plt.show()
    
test_generate_plot_with_style('ggplot',1)
test_generate_plot_with_style('seaborn-colorblind', 2)
test_generate_plot_with_style('grayscale', 3)
```

```python
def test_generate_figure1():
    t = np.arange(0.0, 5.0, 0.01)
    s1 = np.sin(2*np.pi*t)
    s2 = np.sin(4*np.pi*t)
    fig = plt.figure(figsize = (8, 6))
    axes1 = fig.add_subplot(211, title = 'Sin(2*pi*x)')
    axes1.plot(t, s1)
    axes2 = fig.add_subplot(212, title = 'Sin(4*pi*x)', sharex = axes1, sharey = axes1)
    axes2.plot(t, s2)
    plt.savefig('testfigure1.png')
    plt.show()
    
test_generate_figure1()
```

```python
def test_generate_figure2():
    np.random.seed(1000)
    x = np.random.rand(10)
    y = np.random.rand(10)
    z = np.sqrt(x**2 + y**2)
    fig = plt.figure(figsize = (8, 6))
    axes1 = fig.add_subplot(221, title = 'Scatter plot with Upper Triangle Markers')
    axes1.scatter(x, y, s = 80, c = z, marker = '^')
    axes1.set(xticks = [0.0, 0.4, 0.8, 1.2], yticks = [-0.2, 0.2, 0.6, 1.0])
    axes2 = fig.add_subplot(222, title = 'Scatter plot with Plus Markers')
    axes2.scatter(x, y, s = 80, c = z, marker = '+')
    axes2.set(xticks = [0.0, 0.4, 0.8, 1.2], yticks = [-0.2, 0.2, 0.6, 1.0])
    axes3 = fig.add_subplot(223, title = 'Scatter plot with Circle Markers')
    axes3.scatter(x, y, s = 80, c = z, marker = 'o')
    axes3.set(xticks = [0.0, 0.4, 0.8, 1.2], yticks = [-0.2, 0.2, 0.6, 1.0])
    axes4 = fig.add_subplot(224, title = 'Scatter plot with Diamond Markers')
    axes4.scatter(x, y, s = 80, c = z, marker = 'd')
    axes4.set(xticks = [0.0, 0.4, 0.8, 1.2], yticks = [-0.2, 0.2, 0.6, 1.0])
    plt.tight_layout()
    plt.savefig('testfigure2.png')
    plt.show()
    
test_generate_figure2()
```

```python
def test_generate_figure3():
    x = np.arange(1, 101)
    y1 = x; y2 = x**2; y3 = x**3
    fig = plt.figure(figsize = (8, 6))
    g = matplotlib.gridspec.GridSpec(2,2)
    axes1 = fig.add_subplot(g[0,:1], title = 'y = x')
    plt.plot(x, y1)
    axes2 = fig.add_subplot(g[1,:1], title = 'y = x**2')
    plt.plot(x, y2)
    axes3 = fig.add_subplot(g[1:], title = 'y = x**3')
    plt.plot(x, y3)
    plt.tight_layout()
    plt.savefig('testfigure3.png')
    plt.show()
    
test_generate_figure3()
```
