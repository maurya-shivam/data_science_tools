### subplot( nrows, ncols, index )           
##### like  2x2 matrix and placing values using indexes  
'index' is the position in a virtual grid with 'nrows' and 'ncols'  
'index' number varies from 1 to `nrows*ncols`

```python
fig = plt.figure(figsize = (8, 6))
ax1 = fig.add_subplot(2,2,1, title = 'Plot_1')
ax2 = fig.add_subplot(2,2,2, title = 'Plot_2')
ax3 = fig.add_subplot(2,2,3, title = 'Plot_3')
ax4 = fig.add_subplot(2,2,4, title = 'Plot_4')
plt.show()
```

```python
plt.figure(figsize = (15,15))
x = np.linspace(0,2, 23) 
plt.subplot(331, title = 'Line graph')    #   (  these values depends on the fig numbers (2 for 2 fig together, 3 for 3 fig.....works somewhat like index values  (3,3,_) -->  1, 2, 3  space.) )
plt.plot(x, x**3, label = 'Cubic line path');   plt.legend()
plt.subplot(332, title = 'Scatter graph')
plt.scatter(x, x**3, label = 'Cubic path', c = np.random.randint(0, 50, 23), s = np.random.randint(0, 200, 23), edgecolor = 'black', marker = random.choice(['o', 's', '^']));   plt.legend()
plt.subplot(333)
plt.plot(x, x**3, label = 'Cubic path', linestyle = '', marker = 's', markersize = 8, c = 'orange', linewidth = 100) ;plt.title('Scatter graph using plt.plot');   plt.legend()
plt.suptitle('Categorical Plotting')
plt.tight_layout()
plt.show()
```

```python
data = {'a': np.arange(50), 'c': np.random.randint(0, 50, 50), 'd': np.random.randn(50), 'm': random.choice(['s', '^', 'o', 'x', 'd', '<', '>'])}
data['b'] = data['a'] + 10 * np.random.randn(50)
data['d'] = np.abs(data['d']) * 100
fig = plt.figure(figsize = (10, 10))

plt.subplot(221)
#plt.scatter(data['a'], data['b'], c=data['c'], s=data['d'], marker = data['m'])
plt.scatter('a', 'b', c = 'c', s = 'd', marker = data['m'], data = data, alpha = 0.75)
plt.grid(True)

plt.subplot(222)
plt.scatter('a', 'b', c = 'c', s = 'd', marker = data['m'], data = data, alpha = 0.75)
plt.axis([0, 60, 0, 70])     #  [xmin, xmax, ymin, ymax]     #  Axis method overrides the current scale and values get plotted according to that only
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
fig = plt.figure(figsize = (15, 10))
ax = plt.subplot(221)
ax.set(title='Avg. Quarterly Sales', xlabel='Quarter', ylabel='Sales (in millions)')
quarters = np.arange(3);   sales_2017 = [25782, 35783, 36133];   sales_2016 = [28831, 30762, 32178]; sales_2017 = [25782, 35783, 36133]

plt.bar(quarters-0.2, sales_2016, color = 'red', edgecolor = 'g', width = 0.4, label = '2016', align = 'center', linewidth = 5)
plt.bar(quarters+0.2, sales_2017, color = 'orange', edgecolor = 'g', width = 0.4, label = '2017', align = 'center')

plt.xticks(quarters, ['Q1-2017', 'Q2-2017', 'Q3-2017'])
plt.legend(loc = 2)

#   barh
ax = plt.subplot(222)
ax.set(title='Avg. Quarterly Sales', xlabel='Quarter', ylabel='Sales (in millions)')
quarters = np.arange(3);   sales_2017 = [25782, 35783, 36133];   sales_2016 = [28831, 30762, 32178]; sales_2017 = [25782, 35783, 36133]

plt.barh(quarters, sales_2016, color = 'c', edgecolor = 'g', label = '2016', align = 'center', height = 0.4)
plt.yticks(quarters, ['Q1-2017', 'Q2-2017', 'Q3-2017'])
plt.legend(loc = 2)
plt.show()
```

```python
fig = plt.figure(figsize=(6,6))
ax = fig.add_subplot(111)
ax.set(title='Avg. Quarterly Sales')
sales_2017 = [25782, 35783, 36133]
ax.pie(sales_2017, labels = ['Q1', 'Q2', 'Q3'], colors = list('cgy'), autopct = '%1.1f%%', shadow = True, explode = [0,0,0.2], 
       startangle = 90, counterclock = True, wedgeprops = {'linewidth' : 2}, textprops = {'size': 12, 'color': 'k'})       # 'fmt % pct'
plt.legend(loc = 2, fontsize = 12)
plt.show()
```

```python
xval = np.arange(0.1, 4, 0.5);    yval = np.exp(-xval)
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111)
plt.errorbar(xval, yval, xerr = 0.2, yerr = 0.2, ecolor = 'k')
plt.errorbar(xval-2, yval, xerr = 0.2, yerr = 0.2, uplims = True)
plt.errorbar(xval-4, yval, xerr = 0.2, yerr = 0.2, lolims = True)
plt.errorbar(xval-6, yval, xerr = 0.2, yerr = 0.2, uplims = [1,0]*4, lolims = [0, 1]*4)
plt.title('matplotlib.pyplot.errorbar() function Example')
plt.show()
```

```python
np.random.seed(100)
x = np.random.normal(loc = 60, scale = 10, size = 1000)
#x = 60 + 10 * np.random.randn(1000)
fig = plt.figure(figsize = (8, 6))
ax = fig.add_subplot(111)
ax.set(title="Distribution of Student's Percentage", ylabel='Count', xlabel='Percentage')
plt.hist(x, color = 'c', bins = 30, density = True, histtype='bar', align='mid', orientation='vertical')
plt.show()
```

```python
np.random.seed(100)
x = 50 + 10*np.random.randn(1000)
y = 70 + 25*np.random.randn(1000)
z = 30 + 5*np.random.randn(1000)
print(f"{x.max()}   {x.mean()}   {x.min()}")
print(f"{y.max()}   {y.mean()}   {y.min()}")
print(f"{z.max()}   {z.mean()}   {z.min()}")
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111)
ax.set(title="Box plot of Student's Percentage", xlabel='Class', ylabel='Percentage')
ax.boxplot([x, y, z], labels = ['A', 'B', 'C'], notch = True, bootstrap = 100, vert = True)
plt.show()
```
